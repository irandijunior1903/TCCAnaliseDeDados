# Filtrando os dados das escolas públicas da Paraíba (2019)


```python
escolas_br = pd.read_csv('/content/drive/MyDrive/Dados - TCC (Irandi)/divulgacao_ensino_medio_escolas_2019.csv')
```

```python
# Filtrando as colunas necessárias
escolas_colunas_filtradas = ['SG_UF','CO_MUNICIPIO','NO_MUNICIPIO', 'ID_ESCOLA', 'NO_ESCOLA', 'REDE', 'VL_INDICADOR_REND_2019', 
                     'VL_NOTA_MEDIA_2019', 'VL_OBSERVADO_2019', 'VL_PROJECAO_2021']
                     
escolas_br_colunas = escolas_br[escolas_colunas_filtradas]

# Selecionando os dados das escolas públicas na Paraíba
escolas_pb = escolas_br_colunas[(escolas_br_colunas['SG_UF']=='PB') 
                      & (escolas_br_colunas['REDE']=='Estadual')]
```

```python
# Criando uma condição para selecionar os valores do IDEB não nulos
escolas_pb = escolas_pb.loc[(escolas_pb.VL_OBSERVADO_2019.values != '-') 
                                                  & (escolas_pb.VL_PROJECAO_2021.values != '-')]
                                                  
# Resetando o índice
escolas_pb.reset_index(inplace=True, drop=True)                                                
```

```python
# Salvando um arquivo .csv, o qual sera lido do GitHub
escolas_pb.to_csv("dados_escolas_pb_ideb.csv")
```
